## 예제 9-9  스타일 적용하기

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

##  배경 표시 그리기

sns.set_style에는 darkgrid, whitegrid, dark, white, and ticks 총 5가지가 있음

defaultrk darkgred 임

## set_style(style=None, rc=None)
    plot에 미적인 것(aesthetic elements)을 세팅 

    
Parameters
    ----------
    style : dict, None, or one of {darkgrid, whitegrid, dark, white, ticks}
        A dictionary of parameters or the name of a preconfigured set.
    rc : dict, optional
        Parameter mappings to override the values in the preset seaborn
        style dictionaries. This only updates parameters that are
        considered part of the style definition.
        
Examples
    --------
    >>> set_style("whitegrid")
    
    >>> set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
    

## 배경 스타일을 whitegrid로 세팅

```python
sns.set_style("whitegrid")
```

```python

data = np.random.normal(size=(20, 6)) + np.arange(6) / 2

sns.boxplot(data=data)
```

## 배경스타일 white와 ticks 일 때 배경 격자를 없애기

```python
sns.set_style("ticks")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
sns.despine()
```

```python
sns.boxplot(data=data)
sns.despine()
```

## violinplot 과 despine 사용하기

박스그래프를 violinplot 으로 변경하고  despine 함수를 이용해서 축선의 offset을 조정


```python
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True);
```

## whitegrid 에 대한 despine 조정 

## left 조정

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

## right 조정 

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(right=True)
```

## seaborn 에대한 기본 스타일 확인

axes_style(style=None, rc=None)
    Return a parameter dict for the aesthetic style of the plots.
    
Parameters
    ----------
    style : dict, None, or one of {darkgrid, whitegrid, dark, white, ticks}
        A dictionary of parameters or the name of a preconfigured set.
    rc : dict, optional
        Parameter mappings to override the values in the preset seaborn
        style dictionaries. This only updates parameters that are
        considered part of the style definition.
    
    Examples
    --------
    >>> st = axes_style("whitegrid")
    
    >>> set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
    
    >>> import matplotlib.pyplot as plt
    >>> with axes_style("white"):
    ...     f, ax = plt.subplots()
    ...     ax.plot(x, y)               # doctest: +SKIP
