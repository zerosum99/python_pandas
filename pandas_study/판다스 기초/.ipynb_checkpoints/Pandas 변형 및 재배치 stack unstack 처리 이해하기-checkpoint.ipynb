## DataFrame 내에는 stack과 unstack 메소드 처리 기준

###  stack은 열을 행으로 변경 unstack은 행을 열로 변경한다.

```python
from IPython.display import Image
Image(filename='stack_unstack.png') 
```

```python
# Row Multi-Index
row_idx_arr = list(zip(['r0', 'r0'], ['r-00', 'r-01']))
row_idx = pd.MultiIndex.from_tuples(row_idx_arr)

# Column Multi-Index
col_idx_arr = list(zip(['c0', 'c0', 'c1'], ['c-00', 'c-01', 'c-10']))
col_idx = pd.MultiIndex.from_tuples(col_idx_arr)

# Create the DataFrame
d = pd.DataFrame(np.arange(6).reshape(2,3), index=row_idx, columns=col_idx)
d = d.applymap(lambda x: (x // 3, x % 3))

# Stack/Unstack
sx = d.stack()
ux = d.unstack()
```

```python
d
```

### 데이터프레임에서 멀티인덱스로 인덱스와 칼럼이 구성될 때 stack 처리하면

    칼럼에 있는 정보가 인덱스로 이동해서 처리한다.

```python
sx
```

### 데이터 프렘에서 멀티인덱스로 인덱스와 칼럼이 구성될 때 unstack을 처리하면

    인덱스에 있는 정보가 칼럼으로 이동해서 처리한다.

```python
ux
```

###  시리즈에서 unstack 으로 데이터 프레임으로 변경

```python
index = pd.MultiIndex.from_tuples([('one', 'a'), ('one', 'b'),
                                  ('two', 'a'), ('two', 'b')])
```

```python
ss = pd.Series(np.arange(1.0, 5.0), index=index)
```

```python
ss
```

### 데이터 프렘임에서 unstack을 사용해도 시리즈로 변경된다.

```python
df = ss.unstack()
```

```python
df
```

```python
df.unstack()
```

```python
df.unstack().unstack()
```
