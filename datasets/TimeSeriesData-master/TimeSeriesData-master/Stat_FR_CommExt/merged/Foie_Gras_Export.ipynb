```python
# Not sure of this .... need a real fg expert !!!
fg_mask = ((exports['Code_NC8'] == 2074300) | 
           (exports['Code_NC8'] == 2074593) |
           (exports['Code_NC8'] == 2075300) | 
           (exports['Code_NC8'] == 2075593));
fg_exports = exports[fg_mask].copy();
fg_exports['Code_NC8'] = fg_exports['Code_NC8'].apply(str)
```

```python
fg_exports['Month'] = fg_exports.Annee.apply(str) + "-" + fg_exports.Mois.apply(str) + "-01";
fg_exports['Month'] = fg_exports['Month'].apply(lambda x : datetime.datetime.strptime(x, "%Y-%m-%d"))
```

```python
labels1 = pd.read_csv("codes.txt.gz", sep=";", decimal=',');
labels1.columns = "Code_NC8 Label".split()
labels1['Code_NC8'] = labels1['Code_NC8'].apply(str)

fg_exports2 = fg_exports.merge(labels1 , how='left', on='Code_NC8');


def normalize_fg_name_and_variant(full_name):
    lType = "Goose";
    if("Foies gras de canards" in full_name):
        lType = "Duck";

    lCondition = "Forzen"
    if("frais" in full_name):
        lCondition = "Fresh"
    return (lType, lCondition);
    
labels1.head();
```

```python
def normalize_fg_name(name):
    return normalize_fg_name_and_variant(name)[0];

def normalize_fg_variant(name):
    return normalize_fg_name_and_variant(name)[1];




fg_exports2['fg_type'] =fg_exports2['Label'].apply(normalize_fg_name)
fg_exports2['fg_condition'] =fg_exports2['Label'].apply(normalize_fg_variant)
```

```python
country_values_10_best = country_values.sort_values(by='EUROS', ascending=False).head(10)
```

```python
Best_Customers = list(country_values_10_best.index)
```

```python
Best_Customers
```

```python

def reduce_countries(cntry_name):
    if(cntry_name in Best_Customers):
        return cntry_name;
    return "OTHER";

wexp4['Country_1'] = wexp4['Country'].apply(reduce_countries);
```

```python
wexp4['composite_key_1'] = wexp4['fg_type'] + "_" + wexp4['fg_condition'] + "_" + wexp4['Country_1'];
```

```python
df_euros_1 = pd.pivot_table(wexp4, values="EUROS", index=["Month"], columns=["composite_key_1"], fill_value=0) 
df_kgs_1 = pd.pivot_table(wexp4, values="KG", index=["Month"], columns=["composite_key_1"], fill_value=0) 
df_euros_1.head()
```
